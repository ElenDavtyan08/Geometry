# Թեմա 1. Գաղափար եռանկյունների մասին

```python
# 1. Արդյո՞ք եռանկյունը գոյություն ունի։
a = int(input('ներմուծեք կողմի երկարությունը։ ')) # նեմուծել եռանկյան կողմերը
b = int(input('ներմուծեք կողմի երկարությունը։ '))
c = int(input('ներմուծեք կողմի երկարությունը։ '))

if (a+b>c) and (a+c>b) and (c+b>a):
    print('Եռանկյունը գոյություն ունի')
else:
    print('եռանկյունը գոյություն չունի')
    
# 2.Տրված է ուղղանկյուն եռանկյուն, որի սուր անկյուններից մեկի մեծությունը 72° է: Որոշիր այդ եռանկյան մյուս սուր անկյան մեծությունը:
a = int(input('Ներմուծեք անկյան աստիճանային չափը։ '))
b = 90
c = 180 - (b + a)
print('Երրորդ անկյունը հավասար է', c)

# 3.Եռանկյան երկու կողմերը հավասար են, իսկ երրորդ կողմը փոքր է դրանցից 2 անգամ։Գտեք եռանկյան կողմերը, եթե դրա պարագիծը 55 դմ է։
perimeter = int(input("Ներմուծեք պարագծի չափը։ "))
x = 2 * perimeter / 5  # x-ը ստանալու համար անհրաժեշտ է կազմել հետևյալ հավասարումը․ 2x + (x/2) = 55
print(f"Կողմերն են՝ {x}, {x}, {x/2}")

# 4. Եռանկյան երեք կողմներն հարաբերում են ինչպես 2։4։5։ Գտեք եռանկյան կողմերը, եթե դրա պարագիծը 66 սմ է։
a_ = int(input("Ներմուծեք առաջին թիվը։ "))
b_ = int(input("Ներմուծեք երկրորդ թիվը։ "))
c_ = int(input("Ներմուծեք երրորդ թիվը։ "))
perimeter = int(input("Ներմուծեք պարագծի չափը։ "))
x = perimeter //(a_+b_+c_)
print(f"Կողմերն են {x*a_}, {x*b_}, {x*c_}")

#5. Որոշիր NLM եռանկյան անկյունների մեծությունները, եթե ∠N:∠L:∠M = 4:3:5։
n = int(input("Ներմուծեք առաջին թիվը։ "))
l = int(input("Ներմուծեք երկրորդ թիվը։ "))
m = int(input("Ներմուծեք երրորդ թիվը։ "))
whole_parts = n + l + m
angles = 180
N = angles * n // whole_parts
L = angles * l // whole_parts
M = angles * m // whole_parts
result = f"Անկյուններն են` ∠N={N}°, ∠L={L}°, ∠M={M}°"
print(result)
```

# Թեմա 2․ Եռանկյունների տեսակներն ըստ անկյունների

```python
# 1. Հայտնի է, որ՝∠MLK=51°-ի: MLK-ն եռանկյան անկյուններից է: Նշիր MLK եռանկյան տեսակը:
mlk = 51
choice = int(input('Ընտրիր եռանկյունու տեսակը․ 1․սուրանկյուն, 2․բութանկյուն, 3․ուղղանկյուն, 4․հնարավոր չէ որոշել: ')) #ընտրել համարը
if choice == 4:
    print('Ճիշտ է')
else:
    print('Սխալ է')
    
# 2.Եռանկյան արտաքին անկյունը հավասար է 25°։ Ստուգել եռանկյունը սուրանկյուն է, ուղղանկյուն թե՞ բութանկյուն։
external_angle = int(input("Ներմուծեք արտաքին անկյան աստիճանային չափը։ "))
c = 180 - external_angle # ներքին անկյան աստիճանային չափը
if c > 90:
    print('Եռանկյունը բութանկյուն է')
elif c == 90:
    print('Եռանկյունն ուղղանկյուն է')
else:
    print('Եռանկյունը սուրանկյուն է')

# 3.Եռանկյունն ունի 21 աստիճանի երկու անկյուն: Տրված եռանկյունը՝ ուղղանկյուն է, բութանկյուն է, սուրանկյուն է։
a = int(input('Ներմուծեք անկյան աստիճանային չափը։ '))
b = int(input('Ներմուծեք անկյան աստիճանային չափը։ '))
c = 180-(a+b)
if c > 90:
    print('Եռանկյունը բութանկյուն է')
elif c == 90:
    print('Եռանկյունն ուղղանկյուն է')
else:
    print('Եռանկյունը սուրանկյուն է')
    
#4. Տրված է ACB հավասարասրուն ուղղանկյուն եռանկյուն, որի մեջ ∠ABC=45°: Ներքնաձիգին տարված է CD ուղղահայացը, որը 8սմ է։ Գտնել AB-ն։
def count_angle(a, b = 90):
    c = b-a
    return c

# Դիտարկենք ուղղանկյուն △CDB: ∠DBC=45° => ∠DCB=90°-45°=45° => △CDB հավասարասրուն ուղղ․ եռանկյուն է => DB=CD=8
dbc = 45
cdb = 90
dcb = count_angle(dbc, cdb)
db = 8
cd = db

# Դիտարկենք △ACB, որը ուղղանկյուն է։ ...
acd = 90 - dcb

# Դիտարկենք △CDA, որը ուղղանկյուն է։ ...
cda = 90
cad = count_angle(acd, cda)
ad = cd

ab = ad + db
print(f"AB-ի երկարությունը հավասար է {ab}")
```

# Թեմա 3 Եռանկյունների տեսակներն ըստ կողմերի

```python
# 1. Տրված է RDM հավասարասրուն եռանկյան գագաթի` ∠D = 150°, մեծությունը: Որոշիր հիմքին առընթեր անկյունների մեծությունները:
d = int(input('Ներմուծեք անկյան աստիճանային չափը։ '))
r_m = (180-d)//2
print(f'Հիմքին առընթեր անկյունները` ∠R և ∠M, իրար հավասար են, => անկյունային չափը հավասար է {r_m}°')

# 2.Հավասարասրուն եռանկյան հիմքին առընթեր անկյուններից մեկի մեծությունը 35° է:Որոշիր եռանկյան գագաթի անկյան մեծությունը:
a_b = int(input('Ներմուծեք անկյան աստիճանային չափը։ '))
c = 180-(2*a_b)
print(f'Եռանկյան գագաթի անկյան մեծությունը հավասար է {c}°')

#3. Որոշիր KRM հավասարասրուն եռանկյան անկյունների մեծությունները, եթե հիմքին հանդիպակած R գագաթի անկյան կից անկյունը հավասար է 147°-ի:
r_ = int(input('Ներմուծեք կից անկյան աստիճանային չափը։ '))
r = 180 - r_
k_m = (180-r) / 2
print(f'KRM հավասարասրուն եռանկյան անկյուններ են՝ ∠R = {r}°, ∠K = {k_m}°, ∠M = {k_m}°')

# 4.Եռանկյան անկյունների միջոցով որոշել եռանկյան տեսակն ըստ կողմերի․
a = int(input("∠A = "))
b = int(input("∠B = "))
c = int(input("∠C = "))
if a==b==c:
    print('Եռանկյունը հավասարակողմ է')
elif a == b or a == c or b==c:
    print('Եռանկյունը հավասարասրուն է')
else:
    print('Եռանկյունը տարակողմ է')
```
